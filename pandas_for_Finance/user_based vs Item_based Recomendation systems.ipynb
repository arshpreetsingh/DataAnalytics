# User base Vs Item based Recomendation

## User Based
* User properties are considered as normalised data.
* System tries to find relationship between various users.
* Lots of computation happens due to many to one relationship between user and item.

## Item based 
* Item properties are considered as normalised data.
* System tries to find relationship between various items picked by users.
* Computation is less as compared to User based.
