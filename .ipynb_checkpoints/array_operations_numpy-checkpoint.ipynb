```python
# this file is all about learning Numpy and it's arrays in real life. :D
# this can be also called array processing!
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# here we need to visulise the things in real-time so we will use %matplot-inline
%matplotlib inline
```

```python
# ok now we will generate a list of various number along with all required fields
points = np.arange(-5,5,0.01)
```

```python
# now as we can see we have generated amazing+big python liist, it is not big data but we can consider is somehow ;)

# What is Grid?
# Grid is raw shape defined by set of points in the matrix form.

# let's generate grid from various points
dx,dy=np.meshgrid(points,points)
```

```python
# now we will plot it
plt.imshow(dx)
plt.colorbar()
```

```python
# same as we plot dy
plt.imshow(dy)
plt.colorbar()
```

```python
# as we can see that both dx and dy has same set of values but one is horizontal array and other is vertical, this is
#mesh-grid funtion.
z = np.sin(dx)+np.sin(dy)
```

```python
plt.imshow(z)
plt.colorbar()
```

```python
# now we are comparing one of the more amazing method of numpy which is based on the boolean statement for finding values
#in the array
A = np.array([1,2,3,4])
B = np.array([100,200,300,400])
condition = [True,True,False,False]
```

```python
from itertools import izip
answer = [(A_val if cond else None)for A_val,cond in izip(A,condition)]
```

```python
answer
```

```python
# here in the above we got answer for one to one array comparison for A and conditon now same let's try using generator
#express
answer = ((A_val if cond else None)for A_val,cond in izip(A,condition))
```

```python
answer
```

```python
list(answer)
```

```python
#so here we have answer using generator expression 
# now using two lists
answer = [(A_val if cond else B_val)for A_val,B_val,cond in izip(A,B,condition)]
answer


```

```python
#to but we are here to use amazing power of numpy so why to do that?
answer = np.where(condition,A,B)
```

```python
answer
```

```python
# tada where() method in np in much more sexier than any-one another around, What it actually does?
#np.where(condition,A,B) look for A in condition if true otherwise go for
from numpy.random import randn
```

```python
arr = randn(5,5,5)
arr
    
```

```python
# now we have go array of any five random numbers, Okay now what to do next?
# we can define array in any dimension we need just randn(5), randn(5,5) or randn(5,5,5)

# ok now demonstrate on more amazing function of np.where function

if_cond = np.where(arr<0,0,arr)
```

```python
if_cond
```

```python
# so now this file denotes that if any value of arr is (arr<0) we will put 0 there otherwise go with arr value
# let's try with one more example and see if it works
else_condition = np.where(arr<0|arr<1,1,arr)
else_condition
```
